### Generate random numbers

```python
import numpy as np
import matplotlib.pyplot as plt

import seaborn as sns
plt.style.use('seaborn-white')
plt.rcParams['font.family']=['serif']
plt.rcParams['font.size'] = 14
plt.rcParams['xtick.labelsize']=12
plt.rcParams['ytick.labelsize']=12
```

```python
np.random.seed(10)
N = 10000
n = np.arange(N)

ran1 = np.random.uniform(0,1,N) # return random floats in the half-open interval [a, b)
fig, (ax1, ax2) = plt.subplots(figsize=(10,5), ncols=2)
ax1.hist(ran1)
ax2.plot(n, ran1,'.');
```

```python
np.random.seed(10)
ran2 = np.random.random(N) # return random floats in the half-open interval [0.0, 1.0)
fig, (ax1, ax2) = plt.subplots(figsize=(10,5), ncols=2)
ax1.hist(ran2)
ax2.plot(n, ran2,'.');
```

```python
from random import sample, seed

seed(10)

N = 1000
n = np.arange(N)

ran_sample = sample(sorted(ran1), N) # returns random selected elements from sorted list
fig, (ax1, ax2) = plt.subplots(figsize=(10,5), ncols=2)
ax1.hist(ran_sample)
ax2.plot(n, ran_sample,'.');
```
